### 4. Анализ причин повышенного оттока пожилых клиентов

Посмотрим на распределение в зависимости от того, является ли пользователь пожилым 

```python
q = """
SELECT seniorcitizen, churn, COUNT(*) AS count
FROM churn_data
GROUP BY seniorcitizen, churn;
"""

#получаем данные
senior_data = pd.read_sql(q, engine)

#преобразуем данные в удобный формат для построения графика
senior_data = senior_data.pivot(index='seniorcitizen', columns='churn', values='count')


#строим диаграмму
senior_data.plot(kind='bar', stacked=True)
plt.title('Churn Distribution by senior')
plt.xlabel('Seniorcitizen')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

Проведем тест хи- квадрат для проверки нашей гипотезы. Все условия для проведения теста выполнены. 

```python
from scipy.stats import chi2_contingency

print (senior_data)
chi2, p, dof, expected = chi2_contingency(senior_data)
print(chi2, p)
```

Можно заметить, что пенсионеры уходят значительно чаще других пользователей. 
Посмотрим на категориальные данные, чтобы найти связь между тем, что человек-пенсионер и присутствием той или иной услуги. 

В случаях, когда категорий 3, допустим (Yes, No, No phone service) объединяем (No, No phone service) и в итоге получаем ('Yes', 'No')

В качестве p примем за значимость значение < 0.05

```python
query = """
SELECT 
    seniorcitizen,
    -- Преобразование колонок
    CASE 
        WHEN partner IN ('Yes', 'No') THEN partner 
        ELSE 'No' 
    END AS partner,
    CASE 
        WHEN dependents IN ('Yes', 'No') THEN dependents 
        ELSE 'No' 
    END AS dependents,
    CASE 
        WHEN phoneservice IN ('Yes', 'No') THEN phoneservice 
        ELSE 'No' 
    END AS phoneservice,
    CASE 
        WHEN multiplelines IN ('Yes', 'No') THEN multiplelines 
        ELSE 'No' 
    END AS multiplelines,
    CASE 
        WHEN internetservice IN ('DSL', 'Fiber optic') THEN 'Yes' 
        ELSE 'No' 
    END AS internetservice,
    CASE 
        WHEN onlinesecurity IN ('Yes') THEN 'Yes' 
        ELSE 'No' 
    END AS onlinesecurity,
    CASE 
        WHEN onlinebackup IN ('Yes') THEN 'Yes' 
        ELSE 'No' 
    END AS onlinebackup,
    CASE 
        WHEN deviceprotection IN ('Yes') THEN 'Yes' 
        ELSE 'No' 
    END AS deviceprotection,
    CASE 
        WHEN techsupport IN ('Yes') THEN 'Yes' 
        ELSE 'No' 
    END AS techsupport,
    CASE 
        WHEN streamingtv IN ('Yes') THEN 'Yes' 
        ELSE 'No' 
    END AS streamingtv,
    CASE 
        WHEN streamingmovies IN ('Yes') THEN 'Yes' 
        ELSE 'No' 
    END AS streamingmovies,
    paperlessbilling
FROM churn_data;
"""

data = pd.read_sql(query, engine)

#список колонок для анализа
columns_to_test = [
    'partner', 'dependents', 'phoneservice', 'multiplelines', 
    'internetservice', 'onlinesecurity', 'onlinebackup', 
    'deviceprotection', 'techsupport', 'streamingtv', 
    'streamingmovies', 'paperlessbilling'
]

results = []

for col in columns_to_test:
    #создание таблицы сопряженности
    contingency_table = pd.crosstab(data['seniorcitizen'], data[col])
    
    #проверка условия ожидаемых частот
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    
    #сохранение результатов
    results.append({
        'Feature': col,
        'Chi2': chi2,
        'p-value': p,
        'Conclusion': 'Significant' if p < 0.05 else 'Not significant'
    })

results_df = pd.DataFrame(results)
print(results_df[['Feature', 'Chi2', 'p-value', 'Conclusion']])
```

```python
import numpy as np

#список признаков
features = ['partner', 'dependents', 'phoneservice', 'multiplelines', 
            'internetservice', 'onlinesecurity', 'onlinebackup', 
            'deviceprotection', 'techsupport', 'streamingtv', 
            'streamingmovies', 'paperlessbilling']

#расчет корреляции Крамера
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min((k-1), (r-1)))

#создание матрицы корреляции
corr_matrix = pd.DataFrame(index=features, columns=['Cramer_V'])

for feature in features:
    # Формируем таблицу сопряженности
    crosstab = pd.crosstab(data[feature], data['seniorcitizen'])
    corr = cramers_v(crosstab.values)
    corr_matrix.loc[feature, 'Cramer_V'] = corr

#сортировка по силе связи
corr_matrix = corr_matrix.sort_values(by='Cramer_V', ascending=False)
print(corr_matrix)
```

Для более подробного анализа следует рассмотреть значимые параметры. Чтобы не загромождать ноутбук однообразным кодом, был вставлен скрин таблицы по связи ключевой метрики ко всем категориальным переменным

Пожилые люди реже имеют иждивенцев, нежели чем более молодые люди. Из этого можно предположить, что они сами чаще являются иждивенцами, что и влияет на то, что они чаще отписываются от компании.

### 6. Влияние трат пользователей

Посмотрим на влияние потраченных денежных средств на лояльность пользователей. 
По изначальным данным с каггла видно, что распределения ненормальные


Поэтому в нашем случае подходит тест Манна-Уитни, так как он не требует нормальности распределения 

```python
#посмотрим на связь от трат за все время 


from scipy.stats import mannwhitneyu
q = """
SELECT totalcharges, churn
FROM churn_data;
"""


df = pd.read_sql(q, engine)

#преобразуем totalcharges в числовой тип и удаляем некорректные значения
df['totalcharges'] = pd.to_numeric(df['totalcharges'], errors='coerce')  # преобразуем в float
df = df.dropna(subset=['totalcharges'])  # удаляем строки с NaN

#разделяем данные на две группы
group_yes = df[df['churn'] == 'Yes']['totalcharges']
group_no = df[df['churn'] == 'No']['totalcharges']

#проводим U-тест
statistic, p_value = mannwhitneyu(group_yes, group_no, alternative='two-sided')

print(f"Mann-Whitney U statistic: {statistic}")
print(f"p-value: {p_value}")
```

Разница есть. Сравним медианы, чтобы понять нарпавление

```python
median_yes = group_yes.median()
median_no = group_no.median()
print(f"Медиана трат у ушедших клиентов (Yes): {median_yes}")
print(f"Медиана трат у оставшихся клиентов (No): {median_no}")

if median_yes < median_no:
    print("Ушедшие клиенты тратили меньше.")
else:
    print("Ушедшие клиенты тратили больше.")
```

Проведем такое же исследование с колонкой monthlycharges

```python
#посмотрим на связь от трат за месяц



q = """
SELECT monthlycharges, churn
FROM churn_data;
"""


df = pd.read_sql(q, engine)

#преобразуем monthlycharges в числовой тип и удаляем некорректные значения
df['monthlycharges'] = pd.to_numeric(df['monthlycharges'], errors='coerce')  # преобразуем в float
df = df.dropna(subset=['monthlycharges'])  # удаляем строки с NaN

#разделяем данные на две группы
group_yes = df[df['churn'] == 'Yes']['monthlycharges']
group_no = df[df['churn'] == 'No']['monthlycharges']

#проводим U-тест
statistic, p_value = mannwhitneyu(group_yes, group_no, alternative='two-sided')

print(f"Mann-Whitney U statistic: {statistic}")
print(f"p-value: {p_value}")
```

И опять разница имеется. Сравним медианы, чтобы понять нарпавление

```python
median_yes = group_yes.median()
median_no = group_no.median()
print(f"Медиана трат у ушедших клиентов (Yes): {median_yes}")
print(f"Медиана трат у оставшихся клиентов (No): {median_no}")

if median_yes < median_no:
    print("Ушедшие клиенты тратили меньше.")
else:
    print("Ушедшие клиенты тратили больше.")
```

### 8. Рекомендации для бизнеса  

##### 1. Форма контракта
Можно внедрить систему мотивации для клиентов с месячной подпиской, например, предложить скидку 15–20% при переходе на годовой или двухгодичный контракт. Это снизит долю уходящих клиентов, так как данные показывают, что краткосрочные подписки связаны с 40% оттоком. 

##### 2. Сегмент пенсионеров
Для снижения оттока среди пенсионеров можно разработать специализированный тариф, включающий базовые услуги (телефонная связь, интернет) по сниженной цене, а также бесплатную упрощенную техподдержку. Поскольку пенсионеры реже имеют иждивенцев и чаще чувствительны к стоимости, такой тариф может повысить их лояльность. Дополнительно полезно запустить образовательные программы: короткие видеоинструкции по использованию услуг или вебинары с пошаговыми руководствами.

##### 3. Анализ трат
Ушедшие клиенты демонстрируют более высокие ежемесячные траты, но меньшие общие расходы за время сотрудничества. Чтобы увеличить жизненный цикл таких клиентов, можно внедрить накопительные бонусы: например, за каждый месяц использования услуг начислять баллы, которые можно обменять на скидки или дополнительные услуги. Для новых клиентов с месячной подпиской эффективно предложить пробный период премиальных опций (онлайн-безопасность, повышенная скорость интернета) на 1–2 месяца.

##### 4. Приоритетные действия для Fiber optic
Клиенты с интернетом типа Fiber optic уходят в 2.8 раза чаще, чем пользователи DSL. Чтобы снизить этот показатель, можно провести аудит качества связи в регионах с высокой концентрацией таких клиентов и предложить временные бонусы, например, бесплатный апгрейд скорости на 1–2 месяца. Для долгосрочного удержания стоит добавить в тарифы гарантированную минимальную скорость соединения, а в случае её нарушения — автоматическую компенсацию (скидка 10% на следующий месяц).

##### 5. Онлайн-безопасность и техподдержка
